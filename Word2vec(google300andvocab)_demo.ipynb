```python
from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

output_string = StringIO()
with open(dir, 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page in PDFPage.create_pages(doc):
        interpreter.process_page(page)

op2=output_string.getvalue()
print(op2)
```

```python
import nltk
from nltk import tokenize
from nltk.tokenize import BlanklineTokenizer
paras=BlanklineTokenizer().tokenize(op2)
```

```python
import re
import numpy 
import gensim
from gensim.parsing.preprocessing import remove_stopwords
def filter_sentence(sentence, stopwords=False):
    sentence = sentence.lower().strip()
    sentence = re.sub(r'[^a-z0-9\s]', '', sentence)
    if stopwords:
          sentence = remove_stopwords(sentence)
    return sentence
```

```python
def get_filtered_sentences(output,stopwords=False):
    #sents=df[["questions"]];
    filtered_sentences=[]
    for i in range (len(paras)):
        #print(index,row)
        filtered=filter_sentence(output[i],stopwords);
        filtered_sentences.append(filtered);
    return filtered_sentences;

filtered_sentences=get_filtered_sentences(paras,stopwords=True)
print(filtered_sentences[-1]);
#print("\n")
```

```python
from gensim.models import Word2Vec
import gensim.downloader as api
v2w_model=None;
try:
    v2w_model = gensim.models.keyedVectors.load("./w2vecmodel.mod")
    print("loaded w2v model")
except:
    v2w_model = api.load('word2vec-google-news-300')
    v2w_model.save("./w2vecmodel.mod")
    print("Saved w2v_1 model")
    
w2vec_embedding_size=len(v2w_model['computer']);
```

```python
def getwordvec(word,model):
    #samp=model['pay'];
    samp=100;
    vec=[0]*(samp);
    try:
        vec=model.wv.get_vector(word,"True");
    except:
        vec=[0]*(samp);
    return (vec)
    
def getPhraseEmbedding(phrase,model):
    samp=getwordvec('language',model);
    vec=numpy.array([0]*100);
    den=0;
    for word in phrase.split():
        den=den+1;
        #vec=vec+numpy.array(model.wv.get_vecattr(word,"True"));
        vec = vec+getwordvec(word, model);
        #vec=model.wv.get_normed_vectors()
    return vec.reshape(1, -1)     
```

```python
# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',op2)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
from nltk.corpus import stopwords
```

```python
#import nltk
#nltk.download('stopwords')
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)

sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
from gensim.models import Word2Vec
v2w_model = Word2Vec(sentences,vector_size=100,min_count=1) 
#model = Word2Vec(sentences,vector_size=100,min_count=1) 
```

```python
words = v2w_model.wv.key_to_index
words
```

```python
sen_embeddings=[];
for sent in filtered_sentences:
    sen_embeddings.append(getPhraseEmbedding(sent,v2w_model));
```

```python
import sklearn
from sklearn.metrics.pairwise import cosine_similarity;

def retrieverAndPrintAnswer(q,model,sentence_embedding,dtf,sentence):
    start_time = time.time()
    question_embedding = getPhraseEmbedding(q,model);
    max_sim=-1;
    index_sim=-1;
    for index,q_embedding in enumerate(sentence_embedding):
        sim=cosine_similarity(q_embedding,question_embedding)[0][0];
        #print(index,sim, sentence[index])
        if sim>max_sim:
            max_sim=sim;
            index_sim=index;
            #print(index,sim, sentence[index])
    print("\n")
    print("Question: ",q)
    print("\n")
    print("Retrieved: ",dtf[index_sim])
    print("\n")
    print("Search took {:.2f} seconds".format(time.time() - start_time))
    #print(index_sim) 
```
