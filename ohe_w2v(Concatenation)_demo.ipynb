```python
import re
import numpy 
import gensim
from gensim.parsing.preprocessing import remove_stopwords
def filter_sentence(sentence, stopwords=False):
    sentence = sentence.lower().strip()
    sentence = re.sub(r'[^a-z0-9\s]', '', sentence)
    if stopwords:
          sentence = remove_stopwords(sentence)
    return sentence
```

```python
def get_filtered_sentences(output,stopwords=False):
    #sents=df[["questions"]];
    filtered_sentences=[]
    for i in range (len(paras)):
        #print(index,row)
        filtered=filter_sentence(output[i],stopwords);
        filtered_sentences.append(filtered);
    return filtered_sentences;

filtered_sentences=get_filtered_sentences(paras,stopwords=True)
print(filtered_sentences[-1]);
#print("\n")
```

```python
def getwordvec(word,model):
    try:
        #vec = model.wv.get_vector(word,"True") + f[word].to_numpy().flatten();
        v1= f[word].to_numpy().flatten();
        v2= model.wv.get_vector(word,"True")
        vec=np.concatenate((v1, v2), axis = None)
    except:
        vec=[0]*(count_words+220);
        vec= np.array(vec)
    return (vec)
    
def getPhraseEmbedding(phrase,model):
    den=0;
    vec=[0]*(count_words+220);
    vec= np.array(vec)
    for word in phrase.split():
        den=den+1;
        vec=vec + getwordvec(word,model).reshape(1,-1);
        #vec=np.array(vec)
    return vec.reshape(1, -1)     
```

```python
# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',op2)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
from nltk.corpus import stopwords
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)
y=sentences
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
z=sentences
for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences,vector_size=220,min_count=1) 
```

```python
sen_embeddings=[];
for sent in filtered_sentences:
    sen_embeddings.append(getPhraseEmbedding(sent,model));
```
